```python
import pickle
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings("ignore")

# Load the trained model from the pickle file
with open('Logistic_Regression.pkl', 'rb') as f:
    clf = pickle.load(f)

# Prompt the user to enter input values for the features

battery_power=float(input("Enter value for battery power in mAh: "))
blue=float(input("Enter value for blutooth (Yes - 1/No - 0): "))
clock_speed= float(input("Enter value for clock speed: "))
dual_sim=float(input("Enter value for dual sim (Yes - 1/No - 0): "))
fc=float(input("Enter value for front camera mega pixels: "))
four_g=float(input("Enter value for 4G (Yes - 1/No - 0): "))
int_memory=float(input("Enter value for internal memory in GBs: "))
m_dep=float(input("Enter value for mobile thickness : "))
mobile_wt=float(input("Enter value for weight: "))
n_cores=float(input("Enter value no of cores: "))
pc=float(input("Enter value for primary camera in MP: "))
px_height=float(input("Enter value for pixel resolution height: "))
px_width=float(input("Enter value for pixel resolution width: "))
ram=float(input("Enter value for RAM in MB: "))
sc_h=float(input("Enter value for screen height of mobile in cm: "))
sc_w=float(input("Enter value for screen width of mobile in cm: "))
three_g=float(input("Enter value for 3G (Yes - 1/No - 0): "))
touch_screen=float(input("Enter value for touch screen (Yes - 1/No - 0): "))
wifi=float(input("Enter value for Has wifi or not: "))

feature_list = [[battery_power, blue, clock_speed, dual_sim,fc,four_g,int_memory,m_dep,mobile_wt,n_cores,pc,px_height,px_width,ram,sc_h,sc_w,three_g,touch_screen,wifi]]
single_pred = np.array(feature_list).reshape(1,-1)
prediction = clf.predict(single_pred)

# Display the predicted target variable to the user
print("The predicted target variable is:", prediction)
```
